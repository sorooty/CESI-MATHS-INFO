# Travail corrigé

```python
import numpy as np
import math
import pandas as pd
import matplotlib.pyplot as plt


liste_note_proba = [[45, 50,53,55,57,60,65,70,73,76,80,85,90],
              [2,6,5,8,10,9,15,10,7,13,5,5,5]]
liste_note_proba
#Calcul de l'espérance
def experience(liste_note_proba):
    esperance = sum(liste_note_proba[0][i] * liste_note_proba[1][i] for i in range(len(liste_note_proba[0]))) / sum(liste_note_proba[1])
    return esperance

print("L'espérance est de :", experience(liste_note_proba))

#graphique distribution des notes
def graphique_distribution_notes(liste_note_proba):
    plt.bar(liste_note_proba[0], liste_note_proba[1], width=2, color='blue', alpha=0.7)
    plt.xlabel('Notes')
    plt.ylabel('Effectif')
    plt.title('Distribution des notes')
    plt.grid(axis='y', alpha=0.75)
    plt.show(block=True)

graphique_distribution_notes(liste_note_proba)

#Calcul de la variance
def variance(liste_note_proba):
    esperance = experience(liste_note_proba)
    variance = sum(liste_note_proba[1][i] * (liste_note_proba[0][i] - esperance) ** 2 for i in range(len(liste_note_proba[0]))) / sum(liste_note_proba[1])
    return variance
print("La variance est de :", variance(liste_note_proba))

#le prof ajoute 5 points à chaque note
for i in range(len(liste_note_proba[0])):
    liste_note_proba[0][i] += 5
print("Nouvelle liste de notes après ajout de 5 points :", liste_note_proba[0])
print("L'espérance après ajout de 5 points est de :", experience(liste_note_proba))
print("La variance après ajout de 5 points est de :", variance(liste_note_proba))
graphique_distribution_notes(liste_note_proba)
```
